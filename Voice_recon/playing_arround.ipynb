```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
signal = np.zeros(500)
```

```python
signal[:100] = np.exp(np.linspace(0, 1, 100))*2
signal[100: 150] = np.log(np.linspace(16, np.exp(1), 50)) * 2
signal[150: 400] = np.sin(np.linspace(0, (2*np.pi), 250) * 10) + 2
signal[400: ] = -1 * np.log(np.linspace(1, 10, 100 )) + 2

plt.plot(signal)

plt.axvline(x = 0, ls='--', c='r')
plt.axvline(x = 100, ls='--', c='b')
plt.axvline(x = 150, ls='--', c='g')
plt.axvline(x = 400, ls='--', c='black')

plt.axvspan(0, 100, facecolor='r', alpha=0.3, label='Attack')
plt.axvspan(100, 150, facecolor='b', alpha=0.3, label='Decay')
plt.axvspan(150, 400, facecolor='g', alpha=0.3, label='Sustain')
plt.axvspan(400, 500, facecolor='black', alpha=0.3, label='Release')

plt.legend()
plt.show()
```

```python


carrier_signal = np.sin(np.linspace(0, 100, 1000) * 5)

amplitude_signal = np.sin(np.linspace(0,100, 1000) * 0.2)*2

modulated_signal = []
for carrier, amp in zip(carrier_signal, amplitude_signal):
    modulated_signal.append(carrier * amp)

modulated_signal = np.array(modulated_signal
                            )
fig, axs = plt.subplots(3)
axs[0].plot(amplitude_signal)
axs[0].set_title("amplitude signal")
# axs[0].ylim(-5, 5)
# axs[0].yticks(np.arange(-5, 6, 1))

axs[1].plot(carrier_signal)
axs[1].set_title("carrier signal")
# axs[1].ylim(-5, 5)
# axs[1].yticks(np.arange(-5, 6, 1))

plt.subplots_adjust(hspace = 1)


axs[2].plot(modulated_signal)
axs[2].plot(amplitude_signal, c='r')
axs[2].set_title("amplitude modulated signal")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Time array
t = np.linspace(0, 1, 1000, False)  # 1 second

# Information signal
info_freq = 5
info_signal = np.sin(2*np.pi*info_freq*t)*2

# Carrier signal
carrier_freq = 50
carrier_signal = np.sin(2*np.pi*carrier_freq*t)

# Frequency modulation
mod_index = 1
mod_signal = np.sin(2*np.pi*carrier_freq*t + mod_index*info_signal*np.pi)
# Plotting
fig, axs = plt.subplots(3)

# Information signal
axs[0].plot(t, info_signal)
axs[0].set_title("Information signal")

# Carrier signal
axs[1].plot(t, carrier_signal)
axs[1].set_title("Carrier signal")

# Modulated signal
axs[2].plot(t, mod_signal)
axs[2].plot(t, info_signal, c='r')
axs[2].set_title("Modulated signal")

plt.subplots_adjust(hspace = 1)
plt.show()
```
